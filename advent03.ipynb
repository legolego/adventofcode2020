```
import numpy as np
```

```
# Using readlines() 
file1 = open('inputs/input03.txt', 'r') 
Lines = file1.readlines() 
  
count = [0,0,0,0,0] #1,3,5,7,12
# Strips the newline character 
for idx, line in enumerate(Lines): 
    line = line.strip()*1000
    
    if line[idx] == '#':
        count[0] = count[0] +  1

    if line[idx*3] == '#':
        count[1] = count[1] +  1

    if line[idx*5] == '#':
        count[2] = count[2] +  1

    if line[idx*7] == '#':
        count[3] = count[3] +  1

    if idx % 2 == 0 and  line[int(idx/2)] == '#':
        count[4] = count[4] +  1
    
print(count)
print(np.prod(count))
```
