```
# Using readlines() 
file1 = open('inputs/input06.txt', 'r') 
Lines = file1.read() 
```

```
Lines01 = Lines.split("\n\n")
print(Lines01)
```

```
sum = 0
for group in Lines01:
    group = group.replace('\n', ' ').replace(" ", "") 
    #print(group, len(set(group)))
    sum += len(set(group)) 
print(sum)

    
```

```
# https://stackoverflow.com/questions/2541752/best-way-to-find-the-intersection-of-multiple-sets
# https://www.quora.com/How-do-I-find-the-common-characters-present-in-a-list-of-strings-in-Python

sum = 0
for group in Lines01[:]:    
    group = group.split('\n')
    #print(group)
    
    sets = [set(x) for x in group] 
    #print(sets)
    
    #print(set.intersection(*sets))

    #print(len(set.intersection(*sets)))

    sum += len(set.intersection(*sets))

    

print(sum)
```
