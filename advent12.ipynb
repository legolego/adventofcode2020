```
# part 1

def trav_ops(ExecRow = 0, currF='E', x=0, y=0):

    if ExecRow == df.shape[0]:
        return "finished: " + str(x) + ', ' + str(y) 

    #print('row:', df.iloc[ExecRow])
    instr = df.iloc[ExecRow][1]
    val = int(df.iloc[ExecRow][2])
    print("Instrs:", instr, val)

    dirs_letters = ['E', 'S', 'W', 'N']
    

    if instr == 'R':
        val = int(val/90)
        newCurrF = dirs_letters[(dirs_letters.index(currF)+val)%4]
        print('RFrom:', currF, 'To:', newCurrF)
        return trav_ops(ExecRow + 1, newCurrF, x, y)

    if instr == 'L': 
        val = int(-val/90)
        newCurrF = dirs_letters[(dirs_letters.index(currF)+val)%4]
        print('LFrom:', currF, 'To:', newCurrF)
        return trav_ops(ExecRow + 1, newCurrF, x, y) 



    if instr == 'N':
        y += val
    if instr == 'S':
        y += -val
    if instr == 'E':
        x += val
    if instr == 'W':
        x += -val

    if instr == 'F':
        instr = currF

        if instr == 'N':
            y += val
        if instr == 'S':
            y += -val
        if instr == 'E':
            x += val
        if instr == 'W':
            x += -val
        
        currF = instr

    print(instr, val, x, y)
    
    return trav_ops(ExecRow + 1, currF, x, y)


trav_ops(ExecRow = 0, currF='E', x=0, y=0)
```

```
df = pd.read_csv('inputs/input12.txt', delim_whitespace=True, header=None)
df.columns = ['orig']
df['instr'] = df['orig'].astype(str).str[0]
df['val'] = df['orig'].astype(str).str[1:]
df
```

```
# part 2

def trav_ops2(ExecRow = 0, shipx=0, shipy=0, wpx=10, wpy=1):

    if ExecRow == df.shape[0]:
        return "finished: " + str(shipx) + ', ' + str(shipy) 

    
    instr = df.iloc[ExecRow][1]
    val = int(df.iloc[ExecRow][2])
    print("Instrs:", instr, val)


    if instr == 'R':
        val = int(val/90)
        for i in range(val):
            wpxtemp = wpx
            wpx = wpy
            wpy = -wpxtemp


    if instr == 'L':
        val = int(val/90)
        for i in range(val):
            wpxtemp = wpx
            wpx = -wpy
            wpy = wpxtemp
    
    if instr == 'N':
        wpy += val
    if instr == 'S':
        wpy += -val
    if instr == 'E':
        wpx += val
    if instr == 'W':
        wpx += -val       

    
    if instr == 'F':
            shipx = shipx + val*wpx
            shipy = shipy + val*wpy            

    print(instr, val, shipx, shipy, wpx, wpy)
    
    return trav_ops2(ExecRow + 1, shipx, shipy, wpx, wpy)


trav_ops2()
```
