```
import pandas as pd
```

```
df = pd.read_csv('inputs/input08.txt', delim_whitespace=True, header=None)
df
```

```

# part 1, return the Accum

def trav_ops(ExecRow = 0, Accum=0):
    instr = df.iloc[ExecRow][0]
    val = df.iloc[ExecRow][1]
    
    #print(instr, val, Accum)

    if ExecRow in been_to:
        print("Been here:", ExecRow, Accum)
        return Accum
    else:
        been_to.append(ExecRow)

    if instr == 'nop':
        trav_ops(ExecRow + 1, Accum)
    
    if instr == 'acc':
        Accum = Accum + val
        trav_ops(ExecRow + 1, Accum)

    if instr == 'jmp':
        trav_ops(ExecRow + val, Accum)



been_to = []

trav_ops(0, 0)
```
