```
Lines01 = Lines.split("\n")
print(Lines01)
```

```
ret_color_list = []

def get_outer_bags_for_color(color):

    for line in Lines01:
        outer_bag, inner_colors = line.split(' bags contain')
        #print(outer_bag, inner_colors)
        if color in inner_colors:
                       
            ret_color_list.append(outer_bag)
            get_outer_bags_for_color(outer_bag)
    
    
    return set(ret_color_list)

         
my_set = get_outer_bags_for_color('shiny gold')

print(len(my_set))
```

```
# part 2

# Using readlines() 
file1 = open('inputs/input07.txt', 'r') 
Lines = file1.read() 

Lines02 = Lines.split("\n")
print(len(Lines02))

pattern = "(\d\s\w+\s\w+)\sbag"


def get_inner_bags_for_color(color, quant = 1):

    
    sumL = 0
    for line in Lines02:        
        
        outer_bag, inner_colors = line.split(' bags contain')
    
        if outer_bag == color: #found shiny gold as outer bag

            reg_match = re.findall(pattern , inner_colors) # inner is: 2 dark red

            print(reg_match, inner_colors)

            if len(reg_match) > 0:
                sumL = 0
                for bag in reg_match:

                    num, adj, clr = bag.split()
                    print(num, quant, adj, clr)

                    sumL += int(num) * quant + get_inner_bags_for_color(adj + ' ' + clr, int(num) * quant)
                    
                return sumL
            else:
                return 0     
    
    return sumL

         
num_bags = get_inner_bags_for_color('shiny gold')

print(num_bags)
```
