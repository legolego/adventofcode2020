```
def seat_changes1(orig_array, count=0):
    rows = orig_array.shape[0]
    cols = orig_array.shape[1]
    
    s = (rows, cols)
    next_arr = np.zeros(s, dtype=int)

    for r in range(rows):
        for c in range(cols):      

            sumAround = 0
            for nr in [r-1, r, r+1]:            
                for nc in [c-1, c, c+1]:                
                    if 0 <= nr <= rows-1 and 0 <= nc <= cols-1 and not(nr == r and nc == c):
                        #print('new:', nr, nc)
                        if orig_array[nr][nc] != -1:
                            sumAround += orig_array[nr][nc]
            #print('cell:', r, c, 'val:', orig_array[r][c], 'Sum:', sumAround)

            if orig_array[r][c] == 0 and sumAround == 0:
                next_arr[r][c] = 1
            elif orig_array[r][c] == 1 and sumAround >= 4:
                next_arr[r][c] = 0
            else:
                next_arr[r][c] = orig_array[r][c]
    count = count + 1
    if not np.array_equal(next_arr, orig_array):
        print(next_arr, count)        
        return seat_changes1(next_arr.copy(), count)
    else:
        print('Done')
        print(next_arr, count)
        return next_arr.copy()
```

```
def seat_changes2(orig_array, count=0):
    rows = orig_array.shape[0]
    cols = orig_array.shape[1]
    s = (rows, cols)
    next_arr = np.zeros(s, dtype=int)

    for r in range(rows):
        for c in range(cols):

            sumAround = 0                    
            
            #row ahead:
            arr_part = orig_array[r, min(c+1, cols):cols] 
            #print('ahead', arr_part)
            # remove floor seats
            arr_part = arr_part[arr_part != -1]            
            if arr_part.size != 0:
                sumAround += arr_part.take(0)
                #print('ahead-first', arr_part.take(0))
            

            #row behind:
            arr_part = orig_array[r, 0:max(c, 0)]            
            #print('behind', arr_part)
            # remove floor seats
            arr_part = arr_part[arr_part != -1]
            if arr_part.size != 0:
                sumAround += arr_part.take(-1)
                #print('behind-last', arr_part.take(-1))
                

            #col up:
            arr_part = orig_array[0:max(r,0), c]            
            #print('up', arr_part)
            # remove floor seats
            arr_part = arr_part[arr_part != -1]
            if arr_part.size != 0:
                sumAround += arr_part.take(-1)
                #print('up-last', arr_part.take(-1))
                

            #col down:
            arr_part = orig_array[min(r+1, rows):rows, c]            
            #print('down', arr_part)
            # remove floor seats
            arr_part = arr_part[arr_part != -1]
            if arr_part.size != 0:
                sumAround += arr_part.take(0)
                #print('down-first', arr_part.take(0)) 
                

            #diag top left to bot right:
            ofst = c-r
            arr_part = np.diagonal(orig_array, offset=ofst)
            #print('diag1', arr_part)
            if ofst >= 0:

                temp_arr_part = arr_part[0:r].copy()
                #print('+diag1up', temp_arr_part)
                # remove floor seats
                temp_arr_part = temp_arr_part[temp_arr_part != -1]             

                if temp_arr_part.size != 0:
                    sumAround += temp_arr_part.take(-1)
                    #print('+diag1up-last', temp_arr_part.take(-1)) 
                
                
                temp_arr_part = arr_part[min(r+1, len(arr_part)):len(arr_part)].copy()
                #print('+diag1down', temp_arr_part)
                # remove floor seats
                temp_arr_part = temp_arr_part[temp_arr_part != -1]                

                if temp_arr_part.size != 0:
                    sumAround += temp_arr_part.take(0)
                    #print('+diag1down-first', temp_arr_part.take(0)) 
                

            else:
                temp_arr_part = arr_part[0:r+ofst].copy()                
                #print('-diag1up',r, ofst, temp_arr_part)
                # remove floor seats
                temp_arr_part = temp_arr_part[temp_arr_part != -1]
                if temp_arr_part.size != 0:
                    sumAround += temp_arr_part.take(-1)
                    #print('-diag1up-last', temp_arr_part.take(-1))  
                
                temp_arr_part = arr_part[min(r+ofst+1, len(arr_part)):len(arr_part)]
                #print('-diag1down', r, ofst, temp_arr_part)
                # remove floor seats
                temp_arr_part = temp_arr_part[temp_arr_part != -1]
                if temp_arr_part.size != 0:
                    sumAround += temp_arr_part.take(0)
                    #print('-diag1down-first', temp_arr_part.take(0)) 
                

            

            c = c
            flip_r = copy.deepcopy(r)
            r = (np.shape(orig_array)[0]-1)-r
            
            #diag bot left to top right:
            ofst = c-r
            arr_part = np.diagonal(np.flipud(orig_array), offset=ofst)
            #print('diag2', arr_part)
            if ofst >= 0:

                temp_arr_part = arr_part[0:r].copy()
                #print('+diag2up', temp_arr_part)
                # remove floor seats
                temp_arr_part = temp_arr_part[temp_arr_part != -1]             

                if temp_arr_part.size != 0:
                    sumAround += temp_arr_part.take(-1)
                    #print('+diag2up-last', temp_arr_part.take(-1)) 
                
                
                temp_arr_part = arr_part[min(r+1, len(arr_part)):len(arr_part)].copy()
                #print('+diag2down', temp_arr_part)
                # remove floor seats
                temp_arr_part = temp_arr_part[temp_arr_part != -1]                

                if temp_arr_part.size != 0:
                    sumAround += temp_arr_part.take(0)
                    #print('+diag2down-first', temp_arr_part.take(0)) 
                

            else:
                temp_arr_part = arr_part[0:r+ofst].copy()                
                #print('-diag2up',r, ofst, temp_arr_part)
                # remove floor seats
                temp_arr_part = temp_arr_part[temp_arr_part != -1]
                if temp_arr_part.size != 0:
                    sumAround += temp_arr_part.take(-1)
                    #print('-diag2up-last', temp_arr_part.take(-1))  
                
                temp_arr_part = arr_part[min(r+ofst+1, len(arr_part)):len(arr_part)]
                #print('-diag2down', r, ofst, temp_arr_part)
                # remove floor seats
                temp_arr_part = temp_arr_part[temp_arr_part != -1]
                if temp_arr_part.size != 0:
                    sumAround += temp_arr_part.take(0)
                    #print('-diag2down-first', temp_arr_part.take(0)) 

            
            r = flip_r

            #print('cell:', r, c, 'val:', orig_array[r][c], 'Sum:', sumAround)

            if orig_array[r][c] == 0 and sumAround == 0:
                next_arr[r][c] = 1
            elif orig_array[r][c] == 1 and sumAround >= 5:
                next_arr[r][c] = 0
            else:
                next_arr[r][c] = orig_array[r][c]
    count = count + 1

    if not np.array_equal(next_arr, orig_array):
        #print(next_arr, count)
        #unique, counts = np.unique(next_arr, return_counts=True)
        #part_dict = dict(zip(unique, counts))
        #print(part_dict)
        print(count)        
        return seat_changes2(next_arr.copy(), count)
    else:
        print('Done')
        print(next_arr, count)
        unique, counts = np.unique(next_arr, return_counts=True)
        part_dict = dict(zip(unique, counts))
        print(part_dict)  
        return next_arr.copy()
```
