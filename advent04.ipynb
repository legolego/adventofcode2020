```
Lines02 = [l.replace('\n', ' ') for l in Lines.split("\n\n")]
Lines02[:5]
```

```
# https://stackoverflow.com/questions/37449492/matching-optional-capture-groups-in-any-order
pattern = """
            (?=(?:.*(byr:(?P<byr>\d{4})))?)
            (?=(?:.*(iyr:(?P<iyr>\d{4})))?)
            #(?=(?:.*(cid:(?P<cid>\d+)))?)
            (?=(?:.*(pid:(?P<pid>\d{9})))?)
            (?=(?:.*(eyr:(?P<eyr>\d{4})))?)
            (?=(?:.*(hgt:(?P<hgt>\d{2,3}(in|cm))))?)
            (?=(?:.*(hcl:(?P<hcl>\#[a-f0-9]+)))?)
            (?=(?:.*(ecl:(?P<ecl>(amb|blu|brn|gry|grn|hzl|oth))))?)
            """
count = 0
for idx, line2 in enumerate(Lines02):

    m = re.search(pattern , line2, re.VERBOSE)
    mDict = m.groupdict()
    #print('m: ', mDict)

    byr = iyr = eyr = hgt = False

    if None not in mDict.values():
        
        byr =  1920 <= int(mDict['byr']) and int(mDict['byr']) <= 2002
        iyr =  2010 <= int(mDict['iyr']) and int(mDict['iyr']) <= 2020
        eyr =  2020 <= int(mDict['eyr']) and int(mDict['eyr']) <= 2030

        if 'in' in mDict['hgt']:
            hgt = 59 <= int(mDict['hgt'][:-2]) and int(mDict['hgt'][:-2]) <= 76
        else:
            hgt = 150 <= int(mDict['hgt'][:-2]) and int(mDict['hgt'][:-2]) <= 193
        
        if all([byr,iyr,eyr,hgt]):
            count += 1

    print("\n\t", Lines02[idx])   
    print(all([byr,iyr,eyr,hgt]), mDict)
    
print(count)
# one extra for some reason, getting 132, but answer is 131



```
