```
#Part 2
df = pd.read_csv('inputs/input10.txt', delim_whitespace=True, header=None)
df.columns = ['val']
vals = df['val'].to_numpy()

builtin = max(vals) + 3

vals = np.append(vals, builtin)
vals = np.append(vals, 0)
v = np.sort(vals)
v
```

```
#Sharon
diff = [str(v[elem]-v[elem-1]) for elem in range(1, len(v))]
diff = ''.join(diff).split('3')

total = 1

for i in diff:
    if len(i)==2:
        total*=2
    if len(i)==3:
        total*=4
    if len(i)==4:
        total *=7

total
```
